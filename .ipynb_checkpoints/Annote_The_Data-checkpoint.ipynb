```python
import cv2
import numpy as np
import json
import os

# ---------- CONFIG ----------
IMAGE_DIR = "data/images"
OUT_DIR = "data/annotations"

# red color thresholds (BGR because OpenCV)
LOWER_RED_1 = np.array([0, 0, 150])
UPPER_RED_1 = np.array([80, 80, 255])

MIN_BOX_AREA = 20   # remove tiny noise
# ----------------------------

os.makedirs(OUT_DIR, exist_ok=True)

def extract_boxes(image_path):
    img = cv2.imread(image_path)
    if img is None:
        raise ValueError(f"Cannot read {image_path}")

    # Create mask for red color
    mask = cv2.inRange(img, LOWER_RED_1, UPPER_RED_1)

    # Clean mask
    kernel = np.ones((3,3), np.uint8)
    mask = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, kernel, iterations=2)

    # Find contours
    contours, _ = cv2.findContours(
        mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE
    )

    boxes = []
    for cnt in contours:
        x, y, w, h = cv2.boundingRect(cnt)
        if w * h >= MIN_BOX_AREA:
            boxes.append([x, y, x + w, y + h])

    return boxes


def main():
    for fname in os.listdir(IMAGE_DIR):
        if not fname.lower().endswith("annotated.bmp"):
            continue

        img_path = os.path.join(IMAGE_DIR, fname)
        boxes = extract_boxes(img_path)

        out_json = {
            "image": fname,
            "boxes": boxes
        }

        json_path = os.path.join(
            OUT_DIR, fname.replace(".bmp", ".json")
        )

        with open(json_path, "w") as f:
            json.dump(out_json, f, indent=2)

        print(f"{fname}: {len(boxes)} boxes extracted")


if __name__ == "__main__":
    main()
```

```python
import json
image_path="/home/imran/AFB Object Detection/data/images/afb101 annotated.bmp"
json_path="/home/imran/AFB Object Detection/data/annotations/afb101 annotated.json"
def visualize(image_path, boxes):
    img = cv2.imread(image_path)
    for x1,y1,x2,y2 in boxes:
        cv2.rectangle(img,(x1,y1),(x2,y2),(0,255,0),1)
    cv2.imshow("check", img)
    cv2.waitKey(0)

with open(json_path) as f:
    data = json.load(f)

boxes = data["boxes"]

visualize(image_path, boxes)
```
