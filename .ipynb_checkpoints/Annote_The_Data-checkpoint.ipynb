```python
import cv2
import numpy as np
import json
import os

# ---------- CONFIG ----------
IMAGE_DIR = "data/images"
OUT_DIR = "data/annotations"

# red color thresholds (BGR because OpenCV)
LOWER_RED_1 = np.array([0, 0, 150])
UPPER_RED_1 = np.array([80, 80, 255])

MIN_BOX_AREA = 20   # remove tiny noise
# ----------------------------

os.makedirs(OUT_DIR, exist_ok=True)

def extract_boxes(image_path):
    img = cv2.imread(image_path)
    if img is None:
        raise ValueError(f"Cannot read {image_path}")

    # Create mask for red color
    mask = cv2.inRange(img, LOWER_RED_1, UPPER_RED_1)

    # Clean mask
    kernel = np.ones((3,3), np.uint8)
    mask = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, kernel, iterations=2)

    # Find contours
    contours, _ = cv2.findContours(
        mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE
    )

    boxes = []
    for cnt in contours:
        x, y, w, h = cv2.boundingRect(cnt)
        if w * h >= MIN_BOX_AREA:
            boxes.append([x, y, x + w, y + h])

    return boxes


def main():
    for fname in os.listdir(IMAGE_DIR):
        if fname.lower().endswith("unannotated.bmp"):
            continue

        img_path = os.path.join(IMAGE_DIR, fname)
        boxes = extract_boxes(img_path)

        out_json = {
            "image": fname,
            "boxes": boxes
        }

        json_path = os.path.join(
            OUT_DIR, fname.replace(".bmp", ".json")
        )

        with open(json_path, "w") as f:
            json.dump(out_json, f, indent=2)

        print(f"{fname}: {len(boxes)} boxes extracted")


if __name__ == "__main__":
    main()
```

```python
import json
image_path="/home/imran/AFB Object Detection/data/images/afb102 annotated.bmp"
json_path="/home/imran/AFB Object Detection/data/annotations/afb102 annotated.json"
def visualize(image_path, boxes):
    img = cv2.imread(image_path)
    for x1,y1,x2,y2 in boxes:
        cv2.rectangle(img,(x1,y1),(x2,y2),(0,255,0),1)
    cv2.imshow("check", img)
    cv2.waitKey(0)

with open(json_path) as f:
    data = json.load(f)

boxes = data["boxes"]

visualize(image_path, boxes)
```

```python
import cv2
import json
import os

image_path = "/home/imran/AFB Object Detection/data/images/afb105 annotated.bmp"
json_path  = "/home/imran/AFB Object Detection/data/annotations/afb105 annotated.json"
out_path   = "/home/imran/AFB Object Detection/debug/afb105_visualized.png"

os.makedirs(os.path.dirname(out_path), exist_ok=True)

def visualize_and_save(image_path, boxes, out_path):
    img = cv2.imread(image_path)

    for x1, y1, x2, y2 in boxes:
        cv2.rectangle(
            img,
            (int(x1), int(y1)),
            (int(x2), int(y2)),
            ( 0, 255, 0),   
            1                   
        )

    cv2.imwrite(out_path, img)
    print(f"Saved visualization to: {out_path}")

# ---- Load JSON boxes ----
with open(json_path) as f:
    data = json.load(f)

boxes = data["boxes"]

# ---- Save visualization ----
visualize_and_save(image_path, boxes, out_path)
```

```python
#Modified Version


import cv2
import numpy as np
import json
import os
from datetime import datetime

# ---------- CONFIG ----------
IMAGE_DIR = "/home/imran/Documents/AFB_Annonation_JSON/2025_09_03_40x_AFB_Positive"
OUT_DIR = "/home/imran/Documents/JSON/2025_09_03_40x_AFB_Positive"

LOWER_RED = np.array([0, 0, 150])
UPPER_RED = np.array([80, 80, 255])

MIN_BOX_AREA = 20
# ----------------------------

os.makedirs(OUT_DIR, exist_ok=True)


def extract_boxes(image_path):
    img = cv2.imread(image_path)
    if img is None:
        raise ValueError(f"Cannot read {image_path}")

    mask = cv2.inRange(img, LOWER_RED, UPPER_RED)

    kernel = np.ones((3, 3), np.uint8)
    mask = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, kernel, iterations=2)

    contours, _ = cv2.findContours(
        mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE
    )

    boxes = []
    for cnt in contours:
        x, y, w, h = cv2.boundingRect(cnt)
        if w * h >= MIN_BOX_AREA:
            boxes.append({
                "bbox": {
                    "xmin": int(x),
                    "ymin": int(y),
                    "xmax": int(x + w),
                    "ymax": int(y + h)
                },
                "class": "AFB",
                "confidence": 1.0
            })

    return boxes


def build_json(fname, boxes):

    afb_status = "positive" if len(boxes) > 0 else "negative"

    data = {
        "image_info": {
            "file_name": fname,
            "patient_id": "PAT_DUMMY_001",
            "slide_id": "SLIDE_DUMMY_001",
            "magnification": "40x",
            "stain": "Ziehl-Neelsen",
            "afb_status": afb_status,
            "afb_grade": "unknown"
        },

        "annotations": boxes,

        "metadata": {
            "annotator": "auto_extracted",
            "source": "red_box_detection",
            "date": datetime.now().strftime("%Y-%m-%d")
        }
    }

    return data


def main():
    for fname in os.listdir(IMAGE_DIR):

        if fname.lower().endswith("unannotated.bmp"):
            continue

        img_path = os.path.join(IMAGE_DIR, fname)

        boxes = extract_boxes(img_path)

        json_data = build_json(fname, boxes)

        json_path = os.path.join(
            OUT_DIR,
            fname.replace(".bmp", ".json")
        )

        with open(json_path, "w") as f:
            json.dump(json_data, f, indent=2)

        print(f"{fname}: {len(boxes)} boxes extracted")


if __name__ == "__main__":
    main()

```
